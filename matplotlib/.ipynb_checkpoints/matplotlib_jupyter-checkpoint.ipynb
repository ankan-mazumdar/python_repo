```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style

x=np.arange(1,11)
x
```

```python
y = 2*x
print(y)
```

# Line Plot

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot([1,2,3],[4,5,1])
```

```python
style.use('ggplot')
x=[5,8,10]
y=[12,16,6]

x2=[6,9,11]
y2=[22,11,15]

plt.bar(x,y,color='k',label='grp one')
plt.bar(x2,y2,color='b',label='grp two')

plt.xlabel('bar number')
plt.ylabel("bar height")
plt.legend()
plt.title('info')
plt.grid(True,color='r')

plt.show
```

```python
style.use('ggplot')
x=[5,8,10]
y=[12,16,6]

x2=[6,9,11]
y2=[22,11,15]

plt.plot(x,y,'g',label='line one',linewidth=5)
plt.plot(x2,y2,'maroon',label='line two',linewidth=5)

plt.legend()
plt.grid(True,color='khaki')

plt.show
```

```python
plt.plot(x,y,color='r',linestyle=':',linewidth=5)
plt.title('x vs y')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
y1=5*x
plt.plot(x,y,color='r',linestyle='-.',linewidth=5)
plt.plot(x,y1,color='b',linestyle='solid',linewidth=5)
plt.title('x vs y')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```

Adding Subplot

```python
plt.subplot(2,2,1) # 2 rows, 2 cols, index=1
plt.plot(x,y,color='r',linestyle='-.',linewidth=5)

plt.subplot(2,2,2)
plt.plot(x,y1,color='b',linestyle='solid',linewidth=5)

plt.subplot(2,2,3)
plt.plot(x,y1,color='y',linestyle='solid',linewidth=5)

plt.show()
```

BAR PLOT

```python
DS={"A":99,"B":50,"S":70} #dictionary{}
names=list(DS.keys())#key stored in list
scores=list(DS.values())#vaues stored in list

plt.bar(names,scores) #values put in bar plot
plt.show
```

```python
DS={"Ankan":99,"Bitan":50,"Sourav":70} #dictionary{}
names=tuple(DS.keys())#keys convrted stored in tuple
scores=tuple(DS.values())#vaues stored in tuple
plt.title("Success scores")
plt.xlabel(names)
plt.ylabel("scoring")
plt.bar(names,scores) #values put in bar plot
plt.grid(True)
plt.show
```

Scatter PLot

```python
plt.barh(names,scores,color='y') #horizontal bar
plt.title("HOR BAR PLOT")
plt.xlabel(names)
plt.ylabel("scoring")
# plt.bar(names,scores) #values put in bar plot
plt.grid(True)
plt.show()
```

SCATTER PLOT

```python
a=[1,4,6,11,5,2,8,3]
b=[20,10,40,30,50,60, 80, 90]

plt.scatter(a,b)
plt.show
```

```python
a=[1,4,6,11,5,2,8,3]
b=[20,10,12,30,25,16, 28, 9]
c=[2,8,6,13,11,5,2,8]

plt.scatter(a,b, marker='*',color='b', s=100)#s=sixe
plt.scatter(c,b, marker='.',color='g', s=1000)
plt.scatter(b,a, marker='+',color='r', s=100)
plt.show
```

```python
a=[1,4,6,11,5,2,8,3]
c=[2,8,6,13,11,5,2,8]

plt.subplot(1,2,1)
plt.scatter(a,b, marker='*',color='b', s=100)#s=sixe

plt.subplot(1,2,2)
plt.scatter(c,b, marker='.',color='g', s=100)

plt.show
```

Pie Chart

```python
fruit=["Apple","Mango","Grapes","Banana"]
quantity=[39,84,23,44]

plt.pie(quantity,labels=fruit)
plt.show
```

```python
fruit=["Apple","Mango","Grapes","Banana"]
quantity=[39,84,23,44]

plt.pie(quantity,labels=fruit,autopct='%0.01f%%',colors=['blue','orange','grey','pink'])
plt.show
```

DoughNut Chart

```python
fruit=["Apple","Mango","Grapes","Banana"]
quantity=[39,84,23,44]

plt.pie(quantity,labels=fruit,radius=2)
plt.pie([1],colors=['pink'],radius=1)
plt.show
```
